```python
import os
from datetime import datetime

# Input and output folders
input_folder = r"/efs/Incomati/Precipitation/Arc/daily"
output_folder = r"/efs/Incomati/Precipitation/Arc/daily_1"

# Iterate through the files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(".tif"):
        # Extract date information from the filename
        date_str = filename.split("_")[1].split(".")[1]
        print(date_str)
        date_obj = datetime.strptime(date_str, "%Y%m%d")
        
        date_str1 = filename.split(".")[0]
        print(date_str1)

        # Format the date as "YYYY.MM.DD"
        formatted_date = date_obj.strftime("%Y.%m.%d")

        # Create the new filename
        # new_filename = filename.replace(date_str, formatted_date)
        new_filename = f"{date_str1}_{formatted_date}.tif"
        print(new_filename)

        # Construct the full paths
        old_path = os.path.join(input_folder, filename)
        new_path = os.path.join(output_folder, new_filename)

        # Rename the file
        os.rename(old_path, new_path)

print("File renaming completed.")
```
